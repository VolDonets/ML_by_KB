```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
source_file_name = '../Italian_ML_Project/MICC-F220/original_groundtruthDB_220.txt'

with open(source_file_name) as source_file:
    img_names = source_file.readlines()

print(len(img_names))
```

```python
img_cl_names = []
for line in img_names:
    line = line[:-1]
    line = ' '.join(line.split())
    img_cl_names.append([line.split(' ')[0], int(line.split(' ')[1])])
print(img_cl_names)
```

```python
img = cv2.imread('../Italian_ML_Project/MICC-F220/original_img_220/CRW_4809_scale.jpg')
print(img.shape)
plt.imshow(img)
```

```python
img_resized = cv2.resize(img, (224, 224))
print(img_resized.shape)
plt.imshow(img_resized)
```

```python
with open('MICC-F220/micc_f220_classes.txt', 'w') as dest_file:
    for img_data in img_cl_names:
        img = cv2.imread('../Italian_ML_Project/MICC-F220/original_img_220/' + img_data[0])
        img_resized = cv2.resize(img, (224, 224))
        cv2.imwrite('MICC-F220/resized/' + img_data[0].split('.')[0] + '.png', img_resized)
        dest_file.write(img_data[0].split('.')[0] + '.png\t' + str(img_data[1]) + '\n')

dest_file.close()
```
