```python
import cv2
import matplotlib.pyplot as plt

% matplotlib inline
```

```python
from os import listdir
from os.path import isfile, join

mypath = 'MICC-F8multi/original'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
print(onlyfiles)
```

```python
img = cv2.imread('MICC-F8multi/original/DSC_1540_copy_move_Multi.jpg')
print(img.shape)
plt.imshow(img)
```

```python
for img_path in onlyfiles:
    img = cv2.imread('MICC-F8multi/original/' + img_path)
    img_resized = cv2.resize(img, (224, 224))
    cv2.imwrite('MICC-F8multi/resized/' + img_path.split('.')[0] + '.png', img_resized)
```

```python
with open('MICC-F8multi/micc_f8_names.txt', 'w') as dest_file:
    for img_path in onlyfiles:
        dest_file.write(img_path.split('.')[0] + '.png' + '\n')

dest_file.close()
```
